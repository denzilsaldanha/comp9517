```python
import cv2
import numpy as np
import glob2
import os
```

```python
img_array = []
for filename in glob2.glob('./sequence/*.jpg'):
    print(filename)
    img = cv2.imread(filename)
    height, width, layers = img.shape
    size = (width,height)
    img_array.append(img)
```

```python
files_number_sorted = []
subs = "./sequence/"
extension = ".jpg"
for subs, dirs, files in os.walk('./sequence/'):
    #print(subs)
    for file in files:
        image_number = file[:-4]
        files_number_sorted.append(image_number)
files_number_sorted.sort()
#print(files_number_sorted)
```

```python
sorted_images = []
for img in files_number_sorted:
    img_path = subs + img + extension
    sorted_images.append(img_path)
print(sorted_images)
```

```python
images = []
for filename in sorted_images:
    #print(filename)
    img = cv2.imread(filename)
    height, width, layers = img.shape
    size = (width,height)
    images.append(img)
images
```

```python
out = cv2.VideoWriter('project.avi',cv2.VideoWriter_fourcc(*'DIVX'), 4, size)
 
for i in range(len(images)):
    out.write(images[i])
out.release()
```
